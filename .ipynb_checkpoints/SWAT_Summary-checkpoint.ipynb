```python
def get_SWATOUT(data):
    """compiles the multiple runs of swat-cup into one aggregate of outputs"""
    # area m3 of each subbasin from SWAT8 output
    huc_area = {1:16216467.877773, 2:8378026.959396, 3:3019315.786951, 
            4:9027757.552751, 5:5330435.645301, 6:4929530.587929, 
            7:101009.270658, 8:348964.842841, 9:7376895.506662, 
            10:8158347.072924, 11:3865235.704228, 12:7917438.65085, 
            13:1062554.842307, 14:7674964.212416, 15:140334.160831, 
            16:12312081.230987, 17:2900297.068837, 18:7094574.233541, 
            19:1636941.743852, 20:12382378.806859, 21:38065094.166852, 
            22:19783460.921171, 23:4343224.448167, 24:309378.916984, 
            25:155124.480966, 26:1375500.86299, 27:9427879.566498, 
            28:6605710.201975, 29:4715331.786703, 30:5943364.39899, 
            31:6818690.991201, 32:11520624.502263, 33:10214812.18909, 
            34:32198291.444622, 35:1011484.698841, 36:2285541.300413, 
            37:3175397.205058, 38:65206832.107184, 39:27666.63354, 
            40:88092608.620785, 41:11317909.955955, 42:2203324.439687, 
            43:65793486.338053, 44:21421881.636221, 45:45328974.119477, 
            46:56544308.707263, 47:48700908.774075, 48:998086.388689, 
            49:17574.392336, 50:18666703.891484, 51:10780846.880564, 
            52:727771.310336, 53:1411084.653346, 54:5608754.878994, 
            55:4004786.848484, 56:6838701.401007, 57:1736123.946608, 
            58:8093443.617352, 59:2449539.969543, 60:2854012.049465, 
            61:7346357.769751, 62:38348285.462536}
    
    # swat08 swat cup file
    maindir = "H:/GIS/Ogden_Valley/SWAT8/ResRight.Sufi2.SwatCup/Iterations/"

    df = {}
    run = 0

    # duration of model run after warmup
    dfind = pd.DatetimeIndex(freq='MS',start='1/1/2003',end='12/1/2016')
    yrmolist = [str(dfind[i].year) + str(dfind[i].month).zfill(2) for i in range(len(dfind))]

    aetoutfile = maindir + "IterOut21/Sufi2.Out/S-{:}*".format(data)
    for aet in glob.glob(aetoutfile):
        name = os.path.basename(aet)[:-4]
        dfd = {}
        for j in range(0,20):
            shedline = linecache.getline(aet,j*168+j+1)
            shedname = shedline.strip('\n').strip(' ').strip('\n')
            dfd[shedname] = pd.read_csv(aet, sep="  ", index_col='month',
                                        names=['month',aet], skiprows=j*168+j+1, nrows=168,
                                       engine='python')

            dfd[shedname]['yrmo'] = yrmolist
            dfd[shedname].reset_index(inplace=True)
            dfd[shedname].drop('month',axis=1,inplace=True)
            dfd[shedname].set_index('yrmo',inplace=True)
        df[name.split('_')[-1]] = pd.concat(dfd).unstack(level=1)
        df[name.split('_')[-1]].columns = df[name.split('_')[-1]].columns.droplevel(0)
        df[name.split('_')[-1]] = df[name.split('_')[-1]].T

    aetdfs = pd.concat(df, axis=0)
    
    aetdfs.sort_index(axis=1)
    collist = [str(i) for i in range(1,21)]
    aetdfs.reset_index(inplace=True)
    aetdfs.set_index('yrmo',inplace=True)
    aetdfs = aetdfs.rename(columns={'level_0':'Subbasin'})
    aetdfs['Subbasin'] = aetdfs['Subbasin'].apply(lambda x: int(x),1)
    keep = ['mean', 'median', 'std dev', 'Subbasin',
           '5 percent', '95 percent', 'sem', 'bestfit']
    aetdfs['mean'] = aetdfs[collist].apply(lambda x: np.mean(x)*0.001,1)
    aetdfs['median'] = aetdfs[collist].apply(lambda x: np.median(x)*0.001,1)
    aetdfs['std dev'] = aetdfs[collist].apply(lambda x: np.std(x)*0.001,1)
    aetdfs['5 percent'] = aetdfs[collist].apply(lambda x: np.percentile(x, 0.05)*0.001,1)
    aetdfs['95 percent'] = aetdfs[collist].apply(lambda x: np.percentile(x, 0.95)*0.001,1)
    aetdfs['sem'] = aetdfs[collist].apply(lambda x: sp.sem(x)*0.001,1)
    aetdfs['bestfit'] = aetdfs.loc[:,'6']*0.001
    aetdfs = aetdfs[keep]
    aetdfs['area_m2'] = aetdfs['Subbasin'].apply(lambda x: huc_area[int(x)],1)
    for k in keep:
        if k != 'Subbasin':
            aetdfs[k] = aetdfs[k]*aetdfs['area_m2']*0.000810714
    def parsedate(x):
        return pd.datetime(int(str(x)[:4]),int(str(x)[-2:]),1)

    aetdfs['dtyrmo'] = aetdfs.index.map(parsedate)
    aetdfs['datatype'] = data
    aetdfs.reset_index(inplace=True)
    return aetdfs
```

```python
def summarizedf(df, subbasin='Subbasin'):
    df = df.copy().set_index(['dtyrmo'])
    dfyearly = df.groupby([subbasin,df.index.year]).sum()
    dfavgs = dfyearly.groupby(dfyearly.index.get_level_values(0)).mean()
    dftotals = dfavgs.sum()
    dfmonthly = df.groupby([subbasin,df.index.month]).mean()
    dfmonthly_avg = dfmonthly.groupby(dfmonthly.index.get_level_values(1)).sum()
    return dfyearly, dfavgs, dftotals, dfmonthly, dfmonthly_avg
```

```python

def build_df_from_arcpy(table, index_col=None):
    """
    Build a pandas dataframe from an ArcGIS Feature Class.
    
    Uses the arcpy Search Cursor to loop over a feature class in pandas and
    generate a pandas dataframe. If the dataset is a feature class with a
    geometry it will calculate the length and the area before returning, and
    the geometry will be returned as well-known-text.
    
    :param table: The path to the feature class or table
    :type table: str
    :param index_col: A column to use as the dataframe index. If not supplied
                      for feature classes use the Object ID
    :type index_col: str
    :return: dataframe representation of the feature class. Note this is all in
             memory, so be careful with really big datasets!
    :rtype: pd.DataFrame
    """

    desc = arcpy.Describe(table)
    cursor = arcpy.SearchCursor(table)

    new_data = []

    for row in cursor:
        new_row = {}
        for field in desc.fields:
            new_row[field.aliasName or field.name] = row.getValue(field.name)
        new_data.append(new_row)

    try:

        df = pd.DataFrame(new_data)
        df["SHAPEArea"] = df[desc.shapeFieldName].apply(lambda g: g.area)
        df["SHAPELength"] = df[desc.shapeFieldName].apply(lambda g: g.length)
        df[desc.shapeFieldName] = df[desc.shapeFieldName].apply(lambda g: g.WKT)
    except AttributeError:
        # If this is a table in the datbase or on disk, in ArcGIS it won't have
        # either an OID field, nor a geometry
        pass

    return df
```

```python
def summarizearcdf(df, col, subbasin='Subbasin'):
    
    df = df.copy().set_index(['dtyrmo'])
    dfyearly = df.groupby([subbasin,df.index.year]).sum()
    dfavgs = dfyearly.groupby(dfyearly.index.get_level_values(0))[col].agg({'avg':np.mean,
                                                                       'median':np.median,
                                                                       'std':np.std,
                                                                       'sem':sp.sem})
                                                                                                                       
    dftotals = dfavgs.sum()
    dfmonthly = df.groupby([subbasin,df.index.month]).mean()
    dfmonthly_avg = dfmonthly.groupby(dfmonthly.index.get_level_values(1)).sum()
    return dfyearly, dfavgs, dftotals, dfmonthly, dfmonthly_avg
```

```python
def agg_data(df):
    """combines with subbasin shapefile data and aggregates data by subbasin and geology"""
    dfcomb = pd.merge(sheds,df,left_on="Subbasin", right_index=True, how='left')

    data_cols = df.columns
    adj_data_cols = []
    for col in data_cols:
        dfcomb[col+'_area_adj'] = dfcomb[[col,'area_percentage']].apply(lambda x: x[0]*x[1],1)
        adj_data_cols.append(col+'_area_adj')
    FrkNm = dfcomb.groupby(['Fork','Name']).sum()[adj_data_cols]
    Frk = dfcomb.groupby(['Fork']).sum()[adj_data_cols]
    return FrkNm, Frk
```

```python
def get_tables(searchStr,source,variable, stat='MEAN', mult = 1.0):
    tables = arcpy.ListTables(searchStr)
    fields = arcpy.ListFields(tables[0])
    #for table in prism_tables:
    fieldlist = [field.name for field in fields]

    f = {}
    for table in tables:
        f[table] = pd.DataFrame(arcpy.da.TableToNumPyArray(table,fieldlist))
    g = pd.concat(f)
    g.reset_index(inplace=True)
    return fieldlist, g

def calcvols(searchStr,source,variable, stat='MEAN', mult = 1.0):
    g = get_tables(searchStr, source, variable, stat=stat, mult = mult)[1]
    g['YearMonth'] = g['level_0'].apply(lambda x: str(x)[-9:-3] if str(x)[-3:]=='SUM' else str(x)[-6:],1)
    g.drop(['level_0','level_1','OBJECTID'],axis=1,inplace=True)
    g['SOURCE'] = source
    g['variable'] = variable
    g['volume_m_cubed'] = g[stat]*g['AREA']* mult
    g['volume_acft'] = g['volume_m_cubed']*0.000810714
    g['Month'] = g['YearMonth'].apply(lambda x: x[-2:],1)
    g['Year'] =  g['YearMonth'].apply(lambda x: x[:-2],1)
    #g = g[(~g.YearMonth.str.contains('yr'))]
    #g['dt'] = pd.to_datetime(g.YearMonth,errors='coerce',format='%Y%m')
    return g

#tabplace = 'U:/GWP/Groundwater/Projects/BCM/Data/Zonal.gdb'
tabplace = "H:/GIS/Ogden_Valley/New Folder/Zonal_FkGeo.gdb"
outfold = 'M:/PROJECTS/UBM/'
arcpy.env.workspace = tabplace
```
